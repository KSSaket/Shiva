```python
#Details of class 1 passengers
class_1 = []
count = 0 
for i in range (891):
  if (data.Pclass[i]==1):
    count += 1 
    class_1.append(data.Survived[i])
total_1=sum(class_1)
print("Total number of people in class 1 are: {}" .format(count))
print("Total number of people survived in class 1 are {} " .format(sum(class_1)))
sur_rate1=((total_1/count)*100)
print("The Survival Rate in class 1:",sur_rate1)
```

```python
#Details of class 2 passengers
class_2 = []
count = 0 

for i in range (891):
  if (data.Pclass[i]==2):
    count += 1 
    class_2.append(data.Survived[i])
total_2=sum(class_2)
print("Total number of people in class 2 are: {}" .format(count))
print("Total number of people survived in class 2 are {} " .format(sum(class_2)))
sur_rate2=((total_2/count)*100)
print("The Survival Rate in class 2 is:",sur_rate2)
```

```python
#Detils of class 3 passengers
class_3 = []
count = 0 
for i in range (891):
  if (data.Pclass[i]==3):
    count += 1 
    class_3.append(data.Survived[i])
total_3=sum(class_3)
print("Total number of people in class 3 are: {}" .format(count))
print("Total number of people survived in class 3 are {} " .format(sum(class_3)))
sur_rate3=((total_3/count)*100)
print("Survival Rate in class 3 is:",sur_rate3)
```

```python
#By the above observation we can conclude that the survival rate is very less in class 3 compared to the other 2 classes.
#Class 1 people are rescued first and later the subsequent class people are rescued.hence the high survival rate in class 1.
```

```python
#creating X and Y on SibSp and Parch as they have highest relation.
X = data['SibSp']
y = data['Parch']
```

```python
#spliting our variables into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters
lr.params
```
